- Returns `a / b` normally

- Raises `TypeError` with message `"Inputs must be numeric"` if either input is not a number

````python
def safe_divide(a, b):
    # YOUR CODE HERE
    if not isinstance(a, (int, float)) or not isinstance(b, (int, float)):
        raise TypeError("Inputs must be numeric")

    try:
        return a / b
    except ZeroDivisionError:
        return None

# Tests
assert safe_divide(10, 2) == 5.0
assert safe_divide(5, 0) is None
try:
    safe_divide('x', 2)
    assert False, "Should have raised TypeError"
except TypeError as e:
    assert str(e) == "Inputs must be numeric"
print("1.3 passed ✓")
````

### 4.3 Eigenvalues & Eigenvectors

For the matrix:
```
A = [[4, 1],
     [2, 3]]
```

1. Compute eigenvalues and eigenvectors using `np.linalg.eig`.
2. **Verify** the eigenvector equation: `A @ v = λ * v` for each eigenpair.
3. Plot both eigenvectors as arrows, scaled by their eigenvalues, on a 2D plot.
4. **Explain** in a markdown cell (below): what does it geometrically mean for a matrix to stretch a vector?

````python
A = np.array([[4, 1], [2, 3]], dtype=float)

eigenvalues, eigenvectors = np.linalg.eig(A)  # YOUR CODE HERE

print(f"Eigenvalues: {eigenvalues}")
print(f"Eigenvectors (columns):\n{eigenvectors}")

# Verify Av = λv for each eigenpair
# YOUR CODE HERE
for i in range(len(eigenvalues)):
    v = eigenvectors[:, i]
    lhs = A @ v
    rhs = eigenvalues[i] * v

    print(f"\nEigenpair {i+1}:")
    print("A @ v =", lhs)
    print("λ * v =", rhs)

    assert np.allclose(lhs, rhs), "Eigenvector equation failed!"

# Plot
# YOUR CODE HERE
fig, ax = plt.subplots()

for i in range(len(eigenvalues)):
    v = eigenvectors[:, i] * eigenvalues[i]

    ax.quiver(
        0, 0,
        v[0], v[1],
        angles='xy',
        scale_units='xy',
        scale=1,
        label=f'λ={eigenvalues[i]:.2f}'
    )

ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Eigenvectors scaled by Eigenvalues')
ax.axhline(0, color='k', lw=0.8)
ax.axvline(0, color='k', lw=0.8)
ax.legend()

plt.tight_layout()
plt.show()

print("4.3 passed ✓")
````

**Your Geometric Explanation (4.3):**

Eigenvectors are special vectors whose direction does not change when a matrix transformation is applied.  
The matrix may only stretch, shrink, or flip these vectors, and the amount of scaling is given by the corresponding eigenvalue.

Geometrically:
- A matrix transformation can rotate, stretch, compress, or reflect vectors in space.
- Most vectors change both magnitude and direction after transformation.
- Eigenvectors are the directions that remain unchanged.
- Eigenvalues tell how much the vector is scaled:
  - Eigenvalue > 1 -> vector is stretched
  - 0 < Eigenvalue < 1 -> vector is squished/shrunk
  - Eigenvalue < 0 -> vector is flipped and scaled
  - Eigenvalue = 1 -> vector remains unchanged in length

So, when we say a matrix “stretches” a vector, we mean it changes the vector’s length during the transformation while possibly preserving its direction for eigenvectors.

### 4.4 SVD & Dimensionality Reduction

1. Create a random 4×3 matrix `X` (use `np.random.seed(42)`).
2. Compute its **SVD**: `U, S, Vt = np.linalg.svd(X, full_matrices=False)`.
3. **Reconstruct** `X` from the SVD and verify it matches the original.
4. Perform a **rank-1 approximation** using only the largest singular value/vector — call it `X_approx`.
5. Explain the relationship between SVD components and PCA in the markdown cell below.

````python
np.random.seed(42)
X = np.random.randn(4, 3)

U, S, Vt = np.linalg.svd(X, full_matrices=False)  # YOUR CODE HERE

X_reconstructed = U @ np.diag(S) @ Vt  # YOUR CODE HERE — reconstruct from U, S, Vt
X_approx        = S[0] * np.outer(U[:, 0], Vt[0, :])  # YOUR CODE HERE — rank-1 approximation

print(f"U shape: {U.shape}, S shape: {S.shape}, Vt shape: {Vt.shape}")
print(f"Singular values: {S}")
print(f"Reconstruction error: {np.linalg.norm(X - X_reconstructed):.2e}")
print(f"Rank-1 approximation:\n{X_approx}")

assert np.allclose(X, X_reconstructed, atol=1e-10), "Reconstruction failed!"
print("4.4 passed ✓")
````

**SVD → PCA Connection (4.4):**

> *Explain: what are U, S, Vt in SVD, and how do they relate to principal components? Which matrix contains the directions of maximum variance?*


In Singular Value Decomposition (SVD), a matrix X is decomposed as:

X = U S Vt

where:

- U contains the left singular vectors.
  - These represent directions in the sample/data space.
  - Each column of U describes how strongly each data point contributes to a component.

- S contains the singular values.
  - These indicate the importance or strength of each component.
  - Larger singular values correspond to directions with more variance/information.

- Vt contains the right singular vectors.
  - These represent directions in the feature space.
  - The rows of Vt (or columns of V) are the principal directions/components.

Relationship to PCA:
- PCA finds directions of maximum variance in the data.
- These directions are the principal components.
- In SVD, the principal components are contained in V (or equivalently the rows of Vt).
- The singular values in S are related to how much variance each principal component explains.

Therefore:
- The matrix Vt contains the directions of maximum variance.
- The first row of Vt corresponds to the direction with the greatest variance in the dataset.
- Keeping only the top singular values/components performs dimensionality reduction while preserving the most important structure of the data.

### 5.4 Distribution Testing & Stationarity

1. Generate two samples: `s1` ~ Normal(0, 1), `s2` ~ Exponential(1), both n=200 (seed=0).
2. Run the **Kolmogorov-Smirnov test** (`stats.kstest`) comparing each against a standard normal CDF.
3. Generate a non-stationary time series (linear trend + noise). Run the **Augmented Dickey-Fuller test** (`adfuller` from `statsmodels`) and interpret the result.
4. Difference the series once and re-run ADF to confirm stationarity.

````python
from statsmodels.tsa.stattools import adfuller

np.random.seed(0)
s1 = np.random.normal(0, 1, 200)
s2 = np.random.exponential(1, 200)

# KS test — compare each against N(0,1)
ks_s1 = stats.kstest(s1, 'norm')  # YOUR CODE HERE
ks_s2 = stats.kstest(s2, 'norm')  # YOUR CODE HERE
print(f"KS test s1 (normal):      stat={ks_s1.statistic:.4f}, p={ks_s1.pvalue:.4f}")
print(f"KS test s2 (exponential): stat={ks_s2.statistic:.4f}, p={ks_s2.pvalue:.4f}")

# ADF test on non-stationary series
t = np.arange(200)
ts = 0.05 * t + np.random.normal(0, 1, 200)

adf_result = adfuller(ts)  # YOUR CODE HERE — adfuller(ts)
print(f"\nADF statistic: {adf_result[0]:.4f}")
print(f"ADF p-value:   {adf_result[1]:.4f}")
print(f"Stationary?    {adf_result[1] < 0.05}")

# Difference once and re-run
ts_diff = np.diff(ts)  # YOUR CODE HERE — np.diff
adf_diff = adfuller(ts_diff)  # YOUR CODE HERE
print(f"\nAfter differencing — p-value: {adf_diff[1]:.4f}, Stationary? {adf_diff[1] < 0.05}")
````

````python
def compute_psi(expected, actual, bins=10):
    """
    PSI = sum((actual_pct - expected_pct) * ln(actual_pct / expected_pct))
    Add epsilon=1e-10 to avoid log(0).
    """
    # YOUR CODE HERE
    epsilon = 1e-10

    # Create bins based on expected distribution
    breakpoints = np.linspace(0, 100, bins + 1)
    breakpoints = np.percentile(expected, breakpoints)

    # Histogram counts
    expected_counts, _ = np.histogram(expected, bins=breakpoints)
    actual_counts, _ = np.histogram(actual, bins=breakpoints)

    # Convert to proportions
    expected_pct = expected_counts / len(expected)
    actual_pct = actual_counts / len(actual)

    # Avoid division/log of zero
    expected_pct = np.where(expected_pct == 0, epsilon, expected_pct)
    actual_pct = np.where(actual_pct == 0, epsilon, actual_pct)

    # PSI calculation
    psi = np.sum((actual_pct - expected_pct) * np.log(actual_pct / expected_pct))

    return psi

np.random.seed(1)
train_dist = np.random.normal(50, 10, 1000)    # training distribution
drift_dist = np.random.normal(65, 12, 1000)    # shifted production distribution

psi_value = compute_psi(train_dist, drift_dist)
print(f"PSI: {psi_value:.4f}")
print(f"Shift severity: {'Major' if psi_value > 0.2 else 'Minor' if psi_value > 0.1 else 'Stable'}")

# Plot both distributions
fig, ax = plt.subplots()
# YOUR CODE HERE
ax.hist(train_dist, bins=30, alpha=0.6, density=True, label='Training Distribution')
ax.hist(drift_dist, bins=30, alpha=0.6, density=True, label='Production Distribution')
ax.legend(); ax.set_title('Distribution Shift Visualization')
plt.tight_layout(); plt.show()
````

**Concept Drift vs Covariate Drift & PSI Thresholds (5.5):**

- **Concept drift:** A change in the relationship between input features and the target variable over time, causing the model’s learned patterns to become inaccurate.

- **Covariate drift:** A change in the distribution of input features between training and production data while the target relationship may remain unchanged.

- **PSI < 0.1:** Indicates the distribution is stable with little or no meaningful shift.

- **PSI 0.1–0.2:** Indicates a moderate shift that should be monitored.

- **PSI > 0.2:** Indicates a major distribution shift and possible model degradation.

- **Retraining trigger example:** Retrain the model when PSI consistently exceeds 0.2 or when production performance metrics significantly worsen over time.

````python
total = 10
red, blue, green = 4, 3, 3

p_red   = red / total  # YOUR CODE HERE
p_blue  = blue / total  # YOUR CODE HERE
p_green = green / total  # YOUR CODE HERE

# Joint: P(red first AND blue second)
p_red_then_blue = (red / total) * (blue / (total - 1))  # YOUR CODE HERE

# Conditional: P(blue second | red first)
p_blue_given_red = blue / (total - 1)  # YOUR CODE HERE

# Independence check: are P(B|A) == P(B) ?
independent = (p_blue_given_red == p_blue)  # YOUR CODE HERE — True/False

print(f"P(red)={p_red:.3f}, P(blue)={p_blue:.3f}, P(green)={p_green:.3f}")
print(f"P(red,blue): {p_red_then_blue:.4f}")
print(f"P(blue|red): {p_blue_given_red:.4f}")
print(f"P(blue):     {p_blue:.4f}")
print(f"Independent? {independent}")
````

Independence Check:

### 6.3 Bayes' Theorem

**Scenario — Spam filter:**
- P(Spam) = 0.30 (prior)
- P("free" | Spam) = 0.80 (likelihood)
- P("free" | Not Spam) = 0.05

1. Compute P(Spam | "free") using Bayes' theorem.
2. Implement a simple `naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham)` that returns the posterior P(Spam | word).
3. In the markdown cell, map each term — **prior, likelihood, evidence, posterior** — to the spam example.

````python
p_spam            = 0.30
p_free_given_spam = 0.80
p_free_given_ham  = 0.05
p_ham             = 1 - p_spam

# P(free) — law of total probability
p_free = (p_free_given_spam * p_spam +p_free_given_ham * p_ham)   # YOUR CODE HERE

# Bayes
p_spam_given_free = (p_free_given_spam * p_spam) / p_free  # YOUR CODE HERE

print(f"P('free'): {p_free:.4f}")
print(f"P(Spam | 'free'): {p_spam_given_free:.4f}")

def naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham):
    # YOUR CODE HERE
    prior_ham = 1 - prior_spam

    evidence = (p_word_given_spam * prior_spam +p_word_given_ham * prior_ham)

    posterior = (p_word_given_spam * prior_spam) / evidence

    return posterior

pred = naive_bayes_predict(0.30, 0.80, 0.05)
assert abs(pred - p_spam_given_free) < 1e-9
print(f"naive_bayes_predict: {pred:.4f}")
print("6.3 passed ✓")
````

**Bayes Term Mapping (6.3):**

| Term | Formula | In this example |
|------|---------|-----------------|
| Prior | P(Spam) | Probability that an email is spam before seeing the word "free" = 0.30 |
| Likelihood | P(word \| Spam) | Probability that the word "free" appears in spam emails = 0.80 |
| Evidence | P(word) | Overall probability that the word "free" appears in any email = 0.275 |
| Posterior | P(Spam \| word) | Probability that an email is spam after seeing the word "free" = 0.8727 |

### 6.4 Central Limit Theorem

1. Define a population following an **exponential distribution** (λ=1, size=100,000).
2. Repeatedly draw samples of size `n=30` and record their means — do this 5,000 times.
3. Plot the histogram of sample means.
4. Overlay the **theoretical normal distribution** predicted by the CLT: μ = population mean, σ = population_std / sqrt(n).
5. Run a KS test to confirm the distribution of sample means is approximately normal.

````python
np.random.seed(7)
population = np.random.exponential(scale=1.0, size=100_000)

n_samples = 5000
sample_size = 30
sample_means = np.array([np.random.choice(population, size=sample_size, replace=True).mean()
    for _ in range(n_samples)
])
  # YOUR CODE HERE — array of 5000 sample means

pop_mean = population.mean()
pop_std  = population.std()
clt_std  = pop_std / np.sqrt(sample_size)  # YOUR CODE HERE — predicted std of sample means

print(f"Population mean: {pop_mean:.4f}, Population std: {pop_std:.4f}")
print(f"Sample means mean: {sample_means.mean():.4f}, std: {sample_means.std():.4f}")
print(f"CLT predicted std: {clt_std:.4f}")

# Plot
fig, ax = plt.subplots()
# YOUR CODE HERE — histogram of sample means + theoretical normal overlay
ax.hist(sample_means, bins=40, density=True, alpha=0.6, label='Sample Means')

# Theoretical normal distribution overlay
x = np.linspace(sample_means.min(), sample_means.max(), 300)

ax.plot(x,stats.norm.pdf(x, loc=pop_mean, scale=clt_std),linewidth=2,label='CLT Normal Approximation')
ax.set_title('CLT: Distribution of Sample Means')
plt.tight_layout(); plt.show()

# KS test
ks_result = stats.kstest(sample_means, 'norm', args=(pop_mean, clt_std))
print(f"KS test p-value: {ks_result.pvalue:.4f} → Approximately normal? {ks_result.pvalue > 0.05}")
````

**CLT Reflection (6.4):**

> *In 2–3 sentences: Why does this result matter for statistics and machine learning? What assumption does the CLT justify in many classical statistical tests?*


The Central Limit Theorem is important because it explains why averages of random samples tend to follow a normal distribution, even when the original data is not normally distributed. This allows many statistical methods and machine learning techniques to work reliably on real-world data.

The CLT justifies the normality assumption used in many classical statistical tests, such as t-tests, confidence intervals, linear regression inference, and hypothesis testing.
